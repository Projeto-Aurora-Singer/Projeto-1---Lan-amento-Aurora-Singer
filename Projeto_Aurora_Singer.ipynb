Este projeto simula a análise de condições para lançamento de uma nave, utilizando dados de telemetria e regras de validação.

O objetivo é verificar se todos os parâmetros estão dentro de limites seguros e calcular a autonomia energética para a decolagem.

**Importação de Bibliotecas**

Nesta etapa, são importadas as bibliotecas necessárias para manipulação de dados e geração de valores aleatórios.

```python
import pandas as pd
import random as rd
```

**Leitura do Dataset**

O dataset é carregado a partir de um repositório remoto no GitHub.  
Ele contém dados simulados de telemetria da nave, como temperatura, bateria, pressão e autonomia.



```python
# Leitura do dataset utilizando separador ";"
DataFrame = pd.read_csv('https://raw.githubusercontent.com/Projeto-Aurora-Singer/Projeto-1---Lan-amento-Aurora-Singer/refs/heads/main/dataset_telemetria_sintetico_anomalias_ptbr.csv', sep=';')
```

**Exploração Inicial dos Dados**

Aqui analisamos as primeiras linhas do dataset e verificamos a estrutura geral dos dados, incluindo tipos de variáveis e valores não nulos.

```python
# Mostra as primeiras linhas do dataset
DataFrame.head()
```

```python
# Exibe informações gerais (tipos de dados e valores não nulos)
DataFrame.info()
```

Observação:  Grande parte das colunas está no formato texto (object), sendo necessário tratamento para conversão em valores numéricos.

**Simulação de Lançamento**

Uma linha aleatória do dataset é selecionada para simular um cenário real de lançamento.

Essa linha representa as condições atuais da nave no momento da decisão.

```python
# Seleciona uma linha aleatória do dataset para simular um lançamento
# O índice vai de 0 até 499
lancamento = rd.randint(0, 499)

# Extrai os dados da linha selecionada
infosLancamento = DataFrame.iloc[lancamento]

# Exibe os dados do lançamento selecionado
print(infosLancamento)

# Variável de controle da decisão final (1 = READY, 0 = ABORT)
regraFinal = 1
```

**Validação das Condições de Lançamento**

São aplicadas regras de segurança para determinar se a nave está apta para decolagem.

As verificações incluem:
- Temperatura interna e externa
- Estado de carga da bateria (SOC)
- Pressão do tanque
- Status dos módulos críticos
- Comunicação de telemetria
- Autonomia mínima

```python
# Verifica temperatura interna dentro do limite seguro
if not 18 <= (float(infosLancamento["internal_temp_c"].replace(',', '.'))) <= 35:
    regraFinal = 0

# Verifica temperatura externa
if not -5 <= (float(infosLancamento["external_temp_c"].replace(',', '.'))) <= 30:
    regraFinal = 0

# Verifica estado de carga da bateria (SOC)
if not 60 <= (float(infosLancamento["battery_soc_percent"].replace(',', '.'))):
    regraFinal = 0

# Verifica pressão do tanque
if not 95 <= (float(infosLancamento["tank_pressure_bar"].replace(',', '.'))) <= 145:
    regraFinal = 0

# Verifica se módulos críticos estão operacionais
if 0 == infosLancamento['critical_modules_status']:
    regraFinal = 0

# Verifica status da telemetria
if infosLancamento["telemetry_link_status"] != 1:
    regraFinal = 0

# Verifica autonomia mínima necessária
if not (float(infosLancamento["estimated_autonomy_min"].replace(',', '.'))) >= 45:
    regraFinal = 0

# Tempo fixo necessário para decolagem
tempo_decolagem_min = 8

# Verifica se há autonomia suficiente para a decolagem
if (float(infosLancamento["estimated_autonomy_min"].replace(',', '.'))) < tempo_decolagem_min:
    regraFinal = 0

# Resultado final da decisão
if regraFinal == 0:
    print("DECOLAGEM ABORTADA")
else:
    print("PRONTO PARA DECOLAR")
```

**Resultado da Decisão**

Com base nas regras aplicadas, o sistema define automaticamente se a decolagem deve ser autorizada ou abortada.

 ## ANÁLISE ENERGÉTICA CALCULAR AUTONOMIA INICIAL

A análise energética tem como objetivo avaliar se a nave possui autonomia suficiente para executar a decolagem com segurança, considerando consumo e perdas energéticas.

São considerados:
- Capacidade total da bateria
- Energia disponível
- Consumo estimado
- Perdas energéticas

```python
# Calcula a capacidade total da bateria em kWh
capacidadeTotalKwh = float(infosLancamento['battery_voltage_v'].replace(',', '.')) * float(infosLancamento["battery_capacity_ah"].replace(',', '.')) / 1000
print("O total de energia que a bateria pode armazenar é:", capacidadeTotalKwh)
```

```python
# Calcula a energia disponível com base no estado de carga (SOC)
energyAvailableKwh = capacidadeTotalKwh * float(infosLancamento["battery_soc_percent"].replace(',', '.')) / 100
print("A quantidade de energia disponível atualmente na bateria é (kWh):", energyAvailableKwh)
```

```python
# Calcula o consumo estimado durante a decolagem
takeOffConsumptionKwh = float(infosLancamento["power_load_kw"].replace(',', '.')) * (float(infosLancamento["estimated_autonomy_min"].replace(',', '.')) / 60)
print("O consumo estimado de energia durante a decolagem é (kWh):",takeOffConsumptionKwh)
```

```python
# Ajusta o consumo considerando perdas energéticas
realTakeoffConsumptionKwh = takeOffConsumptionKwh * (1 + (float(infosLancamento['energy_loss_percent'].replace(',', '.'))) / 100)
print("O consumo real de energia considerando perdas é (kWh):", realTakeoffConsumptionKwh)
```

```python
# Calcula a autonomia estimada do sistema (em minutos)
estimatedAutonomyMin = (float(infosLancamento["energy_available_kwh"].replace(',', '.')) / (float(infosLancamento["power_load_kw"].replace(',', '.')) * (1 + (float(infosLancamento["energy_loss_percent"].replace(',', '.')) / 100)))) * 60
print("A autonomia estimada do sistema é (minutos):", estimatedAutonomyMin)
```
